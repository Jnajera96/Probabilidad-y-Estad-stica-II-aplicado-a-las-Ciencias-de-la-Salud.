#**Capítulo 1: Los dos pilares de la ciencia de datos en salud**

En el ámbito sanitario, nos enfrentamos constantemente a la variabilidad biológica. No hay dos pacientes iguales. Para dar orden a esta variabilidad, dividimos la estadística en dos grandes ramas: la Descriptiva y la Inferencial.

**1.1 Estadística Descriptiva: El Retrato de los Datos**

La estadística descriptiva tiene como objetivo resumir, organizar y presentar las características de un conjunto de datos para que sean fáciles de entender. No busca sacar conclusiones más allá de lo que tenemos frente a nosotros; simplemente describe "lo que hay".
¿Qué herramientas utiliza?
Medidas de tendencia central: Media (promedio de glucosa), Mediana (el valor central de la estancia hospitalaria) y Moda (la enfermedad más frecuente en consulta).
Medidas de dispersión: Desviación estándar y Rango (para saber qué tanto varían las edades de los pacientes en un estudio).
Herramientas visuales: Histogramas, diagramas de caja (Boxplots) y tablas de frecuencias.
Ejemplo clínico: Si analizas el dataset de diabetes en Orange y generas un histograma del IMC (BMI), estás haciendo estadística descriptiva. Estás visualizando cómo se distribuye el peso en esa población específica sin intentar predecir el futuro de otros pacientes todavía.

**1.2 Estadística Inferencial: El Salto a la Población**

Esta es la rama más poderosa para un profesional de la salud.

La inferencia nos permite tomar la información de una muestra (un grupo pequeño de pacientes) y extender esos resultados a una población entera.

Aquí es donde manejamos la probabilidad para tomar decisiones bajo incertidumbre.
¿Qué herramientas utiliza?
Pruebas de Hipótesis: Para determinar si un nuevo fármaco es realmente más efectivo que el actual o si la diferencia se debe al azar (p-value).

Intervalos de Confianza: Para estimar, por ejemplo, que la verdadera tasa de éxito de una cirugía está entre el 85% y el 92% con un 95% de certeza.

Modelos de Regresión: Como la regresión logística que realizaste en Orange para predecir la diabetes.

Ejemplo clínico: Cuando usas el Data Sampler para entrenar un modelo con el 80% de los datos y luego lo aplicas al 20% restante, estás entrando en el terreno de la inferencia. Estás intentando que tu modelo "aprenda" patrones generales que sirvan para diagnosticar a pacientes que el sistema aún no conoce.


**1.3 Cálculo del tamaño de la muestra**

**¿Cuántos pacientes son suficientes?**
En la investigación médica, una de las preguntas más frecuentes es:

**¿A cuántas personas debo incluir en mi estudio?**
Si la muestra es muy pequeña, los resultados podrían ser fruto del azar y no ser válidos.

Si es innecesariamente grande, se desperdician recursos y se expone a más pacientes de lo debido a pruebas o tratamientos experimentales.

El cálculo del tamaño de la muestra es el proceso matemático para determinar el número mínimo de unidades necesarias para que los resultados tengan validez estadística y poder clínico.

Para calcular una muestra, necesitamos conocer estos cuatro parámetros estadísticos:
Nivel de Confianza: Es la seguridad que deseamos de que el resultado sea real.

En salud, lo estándar es el 95%. Significa que, si repitiera el estudio 100 veces, en 95 de ellas obtendremos el mismo resultado.

Margen de Error o Precisión (E): Es la cantidad de error que estamos dispuestos a tolerar. Por ejemplo, si estimamos una prevalencia de diabetes del 15% con un error del 3%, aceptamos que el valor real esté entre 12% y 18%.

Variabilidad del Fenómeno (p y q): Se refiere a qué tan común es lo que estudiamos. Si no conocemos la prevalencia previa, solemos usar un enfoque conservador del 50% (0.5), que es donde la variabilidad es máxima.

Poder Estadístico: Es la capacidad del estudio para detectar una diferencia si esta realmente existe. Lo ideal es un poder del 80% o 90%.

import math

def calculate_infinite_sample_size(Z, p, q, E):
    """
    Calcula el tamaño de la muestra para una población infinita.

    Args:
        Z (float): Nivel de confianza (valor Z).
        p (float): Probabilidad de éxito.
        q (float): Probabilidad de fracaso (1 - p).
        E (float): Margen de error.

    Returns:
        float: Tamaño de la muestra para una población infinita.
    """
    N_infinito = (Z**2 * p * q) / (E**2)
    return N_infinito
